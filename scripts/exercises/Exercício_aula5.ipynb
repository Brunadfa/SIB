__RidgeRegression:__
__1.__ Use the dataset cpu.csv;
__2.__ Divide the dataset into train and test sets;
__3.__ Train the model. Which score do you get? And the cost?

```python
import numpy as np
import pandas as pd
from si.Data.dataset import Dataset
from si.IO.CSV import read_csv
from si.models.ridge_regression import RidgeRegression
from si.models.logistic_regression import LogisticRegression
from si.model_selection.split import train_test_split
```

```python
cpu = read_csv(r'C:\Users\bruna\PycharmProjects\SIB\datasets\cpu\cpu.csv', sep=",", label=True, features=True)
```

```python
# standardize data
from sklearn.preprocessing import StandardScaler
cpu.X = StandardScaler().fit_transform(cpu.X)
cpu.X
```

```python
# split dataset
train_dataset, test_dataset = train_test_split(cpu)
test_dataset.y
```

```python
# ridge regressor
ridge = RidgeRegression(max_iter=2000)
ridge.fit(train_dataset)
predictions = ridge.predict(test_dataset)
predictions
```

```python
ridge.score(test_dataset)
```

```python
ridge.cost(test_dataset)
```

```python
ridge.cost_history
```
